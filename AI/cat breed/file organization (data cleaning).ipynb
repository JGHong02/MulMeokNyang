```python
import os
import random
import shutil
import re
```

```python
path = "images/"
# Define a list of image extensions
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
breeds = ['Abyssinian', 'Bengal', 'Birman', 'Bombay', 'British Shorthair', 'Egyptian Mau', 'Maine Coon', 'Persian', 'Ragdoll', 'Russian Blue', 'Siamese', 'Sphynx']
```

```python
imgs_list = [filename for filename in os.listdir(path) if os.path.splitext(filename)[-1] in image_extensions]
len(imgs_list)
```

```python
# All images with 1st letter as captial are cat images
# images with small first letter are dog images

for i in imgs_list:
    if i[0].islower():
        os.remove(path + i)

imgs_list = [filename for filename in os.listdir(path) if os.path.splitext(filename)[-1] in image_extensions]
len(imgs_list)
```

```python
# Create destination folders if they don't exist
for i in breeds:
    if not os.path.exists(os.path.join(path, i)):
        os.makedirs(os.path.join(path, i))
```

```python
# List of breed folders
breed_folders = ['Abyssinian', 'Bengal', 'Birman', 'Bombay', 'British Shorthair',
                 'Egyptian Mau', 'Maine Coon', 'Persian', 'Ragdoll',
                 'Russian Blue', 'Siamese', 'Sphynx']

# Source directory where the images are located
source_directory = "./images/"

# Iterate through files in the source directory
for filename in os.listdir(source_directory):
    # Get the breed from the filename using regular expression
    match = re.match(r'([a-zA-Z]+(?:_[a-zA-Z]+)*)_?\d+\.jpg', filename)
    
    if match:
        breed = match.group(1).replace('_', ' ')
        
        # Check if the breed is valid
        if breed in breed_folders:
            # Create the destination directory if it doesn't exist
            destination_directory = os.path.join(source_directory, breed)
            os.makedirs(destination_directory, exist_ok=True)
            
            # Build the source and destination paths
            source_path = os.path.join(source_directory, filename)
            destination_path = os.path.join(destination_directory, filename)
            
            # Move the file to the corresponding breed folder
            shutil.move(source_path, destination_path)
            
            print(f"Moved {filename} to {breed} folder.")
        else:
            print(f"Invalid breed in filename: {filename}")
    else:
        print(f"Filename does not match expected pattern: {filename}")
```

```python
folders = os.listdir(path)
for i in range(len(folders)):
    imgs_list = [filename for filename in os.listdir(path + folders[i] + '/') if os.path.splitext(filename)[-1] in image_extensions]
    print(folders[i] + ": ", len(imgs_list))
```

We are going to add just 200 images and perform image augmentation to avoid class imbalance problem such as Under-sampling or oversampling.

https://www.quora.com/Does-the-number-of-images-for-each-category-matter-while-training-for-a-convolutional-neural-network


```python
# Set the source directory where the images are located
source_directory = "./images new/"

# Set the destination directory where the selected images will be moved
destination_directory = "./images/"

# Set the number of images to select for each class
images_to_select = 200

# Iterate through each class
for cat_class in breeds:
    # Create the destination directory for the current class
    destination_class_directory = os.path.join(destination_directory, cat_class)
    os.makedirs(destination_class_directory, exist_ok=True)

    # Get a list of all images in the source directory for the current class
    class_images = [img for img in os.listdir(os.path.join(source_directory, cat_class)) if img.endswith('.jpg')]

    # Randomly select 200 images
    selected_images = random.sample(class_images, 200)

    # Move the selected images to the destination directory
    for img in selected_images:
        source_path = os.path.join(source_directory, cat_class, img)
        destination_path = os.path.join(destination_class_directory, img)
        shutil.copyfile(source_path, destination_path)

        print(f"Moved {img} to {destination_class_directory}")
```
