```python
import os
import random
import shutil
```

```python
data_path = "data/"
```

```python
# path to destination folders
train_folder = os.path.join(data_path, 'train')
val_folder = os.path.join(data_path, 'eval')
test_folder = os.path.join(data_path, 'test')
```

```python
# Define a list of image extensions
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
```

```python
# Create a list of image filenames in 'data_path'
imgs_list = [filename for filename in os.listdir(data_path) if os.path.splitext(filename)[-1] in image_extensions]
imgs_list
```

```python
# Sets the random seed 
random.seed(42)
```

```python
# Shuffle the list of image filenames
random.shuffle(imgs_list)
```

```python
# determine the number of images for each set
train_size = int(len(imgs_list) * 0.7)
val_size = int(len(imgs_list) * 0.15)
test_size = int(len(imgs_list) * 0.15)
```

```python
# Create destination folders if they don't exist
for folder_path in [train_folder, val_folder, test_folder]:
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
```

```python
# Copy image files to destination folders
for i,   f in enumerate(imgs_list):
    if i < train_size:
        dest_folder = train_folder
    elif i < train_size + val_size:
        dest_folder = val_folder
    else:
        dest_folder = test_folder
    shutil.copy(os.path.join(data_path, f), os.path.join(dest_folder, f))
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Specify paths
data_path = 'images new/'
train_folder = 'data/train/'
val_folder = 'data/val/'
test_folder = 'data/test'

# Create destination folders if they don't exist
for folder in [train_folder, val_folder, test_folder]:
    os.makedirs(folder, exist_ok=True)

# Get the list of class folders in the data path
class_folders = [f for f in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, f))]
print(class_folders)

for folder in [train_folder, val_folder, test_folder]:
    for classes in class_folders:
        os.makedirs(folder + classes, exist_ok=True)

# Loop through each class folder
for class_folder in class_folders:
    class_path = os.path.join(data_path, class_folder)

    # Get the list of image files in the class folder
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

    # Split the image files into train, val, test sets
    train_files, temp_files = train_test_split(image_files, test_size=0.4, random_state=42)
    val_files, test_files = train_test_split(temp_files, test_size=0.5, random_state=42)

    # Copy images to destination folders
    for file in train_files:
        shutil.copy(os.path.join(class_path, file), os.path.join(train_folder, class_folder, file))

    for file in val_files:
        shutil.copy(os.path.join(class_path, file), os.path.join(val_folder, class_folder, file))

    for file in test_files:
        shutil.copy(os.path.join(class_path, file), os.path.join(test_folder, class_folder, file))
```
